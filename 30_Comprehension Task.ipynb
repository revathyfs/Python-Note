# üß© Part 1: List Comprehension Practice

```python
# 1. Create a list of squares of numbers from 1 to 10.

squares = [x**2 for x in range(1, 11)]
print(squares)
```

```python
# 2. Create a list of even numbers between 1 and 20.

evens = [x for x in range(1, 21) if x % 2 == 0]
print(evens)
```

```python
# 3. Create a list of the first letter of each word in a list.

words = ["apple", "banana", "cherry", "date"]
first_letters = [word[0] for word in words]
print(first_letters)
```

```python
# 4. Replace all vowels in a string with ‚Äò*‚Äô.

text = "comprehension"
vowels = 'aeiou'
masked = ['*' if ch in vowels else ch for ch in text]
print("".join(masked))
```

```python
# 5. Create a list of tuples (number, square) for numbers 1‚Äì5.

pairs = [(x, x**2) for x in range(1, 6)]
print(pairs)
```

# üìò Part 2: Dictionary Comprehension Practice

```python
# 6. Create a dictionary with numbers as keys and their cubes as values.

cubes = {x: x**3 for x in range(1, 6)}
print(cubes)
```

```python
# 7. Convert a list of fruits into a dictionary with fruit name ‚Üí length.

fruits = ["apple", "banana", "kiwi"]
fruit_length = {fruit: len(fruit) for fruit in fruits}
print(fruit_length)
```

```python
# 8. Filter dictionary items where value > 50.

marks = {'John': 45, 'Emma': 75, 'Liam': 60, 'Olivia': 35}
passed = {name: score for name, score in marks.items() if score > 50}
print(passed)
```

```python
# 9. Swap keys and values in a dictionary.

data = {'a': 1, 'b': 2, 'c': 3}
swapped = {v: k for k, v in data.items()}
print(swapped)
```

```python
# 10. Create a dictionary with numbers 1‚Äì5 and label them as ‚Äúeven‚Äù or ‚Äúodd‚Äù.

labels = {x: ("even" if x % 2 == 0 else "odd") for x in range(1, 6)}
print(labels)
```

# üåø Part 3: Set Comprehension Practice

```python
# 11. Create a set of squares for numbers 1‚Äì10.

squares = {x**2 for x in range(1, 11)}
print(squares)
```

```python
# 12. Get all unique vowels present in a string.

text = "comprehension"
vowels = {ch for ch in text if ch in 'aeiou'}
print(vowels)
```

```python
# 13. Create a set of numbers divisible by both 2 and 3 between 1‚Äì30.

divisible = {x for x in range(1, 31) if x % 2 == 0 and x % 3 == 0}
print(divisible)
```

```python
# 14. Extract unique letters from a list of words.

words = ['apple', 'banana']
unique_letters = {ch for word in words for ch in word}
print(unique_letters)
```
